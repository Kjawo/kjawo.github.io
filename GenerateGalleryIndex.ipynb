```polyglot-notebook
#r "nuget: TinyPNG"

using System;
using System.IO;
using TinyPng;

string apiKey = Environment.GetEnvironmentVariable("TINYPNG_API_KEY");
string inputFolder = @"C:\Users\kjawo\Documents\repo\kjawogithubio\gallery\archive\g02\2"; 
string outputFolder = @"C:\Users\kjawo\Documents\repo\kjawogithubio\gallery\archive\g02\2\CompressedImages"; 

var png = new TinyPngClient(apiKey);

Directory.CreateDirectory(outputFolder);
CompressImagesInFolder(inputFolder, outputFolder, png);

static void CompressImagesInFolder(string inputFolder, string outputFolder, TinyPngClient png)
{
    foreach (var file in Directory.GetFiles(inputFolder))
    {
        if (file.EndsWith(".png", StringComparison.OrdinalIgnoreCase) ||
            file.EndsWith(".jpg", StringComparison.OrdinalIgnoreCase) ||
            file.EndsWith(".jpeg", StringComparison.OrdinalIgnoreCase))
        {
            string inputFile = file;
            string outputFile = Path.Combine(outputFolder, Path.GetFileName(file));

            if (File.Exists(outputFile))
            {
                Console.WriteLine($"Skipping {inputFile} as output file already exists.");
                continue;
            }

            try
            {
                CompressImage(inputFile, outputFile, png);
            }
            catch (Exception ex)
            {
                Console.WriteLine($"Error compressing image: {inputFile}. {ex.Message}");
            }
        }
    }
}

static void CompressImage(string inputFile, string outputFile, TinyPngClient png)
{
        var result = png.Compress(inputFile)
            .Download()
            .SaveImageToDisk(outputFile);;

        if (result != null)
        {
            Console.WriteLine($"Image compressed successfully: {inputFile} -> {outputFile}");
        }
        else
        {
            Console.WriteLine($"Error compressing image: {inputFile}");
        }
}
```
